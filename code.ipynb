```python
# This is the main File 
import pandas as pd
import matplotlib.pyplot as plt


# Exploratory data analysis

file_path = 'data/HallOfFame.csv'

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)

# drop non-player inductees
players_df = df[df['category'] == 'Player']

# Encode Y and N values for inducted column
players_df['inducted'] = players_df['inducted'].replace('Y', 1)
players_df['inducted'] = players_df['inducted'].replace('N', 0)

# Group by 'yearid' and sum the 'inducted' column
counts = players_df.groupby('yearid')['inducted'].sum()

print(counts)

# Plot a bar chart
plt.figure(figsize=(15, 6))  # Adjust the width and height as needed
counts.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number of Inducted Players')
plt.title('Count of Inducted Players by Year')

# Show the plot
plt.show()
```

```python
# Combine players' season data into career stats

# get all unique nameIDs
unique_playerids = players_df['playerID'].unique()

print("Unique player IDs:")
print(len(unique_playerids))

# explore batting data
file_path = 'data/Batting.csv'

# Read the CSV file into a Pandas DataFrame
batting_df = pd.read_csv(file_path)

column_names = batting_df.columns
career_df = pd.DataFrame(columns=column_names)


# get data for player
for name in unique_playerids:
    filtered_df = batting_df[batting_df['playerID'] == name]
    # Sum all columns
    column_sums = filtered_df.sum(axis=0, numeric_only=True)
    career_df = career_df.append(column_sums, ignore_index=True)
    career_df.iloc[-1, career_df.columns.get_loc('playerID')] = name

print('done')
print(career_df)
```

```python
# career_df holds career stats of all players
# add hof column
career_df['HOF'] = 0
hof_player_ids = players_df.loc[players_df['inducted'] == 1, 'playerID'].tolist()

# add 1's for each hof player
for name in hof_player_ids:
    career_df.loc[career_df['playerID'] == name, 'HOF'] = 1
```

```python
# Explore career data and hof result

# Print correlation matrix
correlation_matrix = career_df.corr()
print(correlation_matrix)
```
